```python
from dfs1 import dfs1
```

```python
filename = 'dfs1 FilePath'
dfs1_obj = dfs1(filename)
output = dfs1_obj.read_dfs1()
output['2004-02-04 00:00:00']
```

```python
import matplotlib.pyplot as plt
output_data = output['2004-02-04 00:00:00']

ax1 = plt.subplot(3,1,1)
ax1.plot(output_data.loc[:,output_data.columns.values[0]],'k--')
ax1.xaxis.set_visible(False)
ax1.set_ylabel('(m)')
#ax1.yaxis.set_ticks([2.5,3,3.5])
ax1.set_xlabel('Wave Height')
ax1.legend()

ax2 = plt.subplot(3,1,2)
ax2.plot(output_data.loc[:,output_data.columns.values[1]],'k-.')
ax2.xaxis.set_visible(False)
ax2.set_ylabel('(s)')
#ax2.yaxis.set_ticks([8.1,8.3,8.5])
ax2.set_xlabel('Wave Period')
ax2.legend()

ax3 = plt.subplot(3,1,3)
ax3.plot(output_data.loc[:,output_data.columns.values[3]],'k-')
ax3.xaxis.set_visible(False)
ax3.set_ylabel('(deg)')
#ax3.yaxis.set_ticks([210,220,230])
ax3.set_xlabel('Mean Direction')
ax3.legend()
```
